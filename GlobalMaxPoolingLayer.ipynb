```python
import cv2
import tensorflow as tf
from tensorflow.keras.models import Model
from google.colab.patches import cv2_imshow
from tensorflow.keras.layers import Input, GlobalMaxPooling2D
```

```python
!wget https://raw.githubusercontent.com/niazwazir/DEEP_LEARNING_ACADEMY/main/Images/GlobalMaxPooling.png
```

```python
img = cv2.imread('GlobalMaxPooling.png', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

```python
def GlobalPooling_Operation():
    image = Input(shape=(232,232,3))
    AP = GlobalMaxPooling2D()(image)
    model = Model(inputs=image, outputs=AP)
    return model
```

```python
model = GlobalPooling_Operation()
model.summary()
```

```python
tf.keras.utils.plot_model(
    model, to_file='model.png', show_shapes=False, show_layer_names=True, rankdir='LR', expand_nested=False, dpi=96
)
```
