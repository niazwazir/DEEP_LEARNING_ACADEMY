```python
import cv2
import tensorflow as tf
from tensorflow.keras.models import Model
from google.colab.patches import cv2_imshow
from tensorflow.keras.layers import Input, AveragePooling2D
```

```python
img = cv2.imread('AveragePoolingOPeration.png', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

```python
def Average_Operation():
    image = Input(shape=(232,232,3))
    AP = AveragePooling2D(pool_size=(2,2))(image)
    model = Model(inputs=image, outputs=AP)
    return model
```

```python
model = Average_Operation()
model.summary()
```

```python
tf.keras.utils.plot_model(
    model, to_file='model.png', show_shapes=False, show_layer_names=True, rankdir='LR', expand_nested=False, dpi=96
)
```

```python
def Average_Operation():
    image = Input(shape=(232,232,3))
    AP = AveragePooling2D(pool_size=(2,2))(image)
    AP1 = AveragePooling2D(pool_size=(2,2))(AP)
    model = Model(inputs=image, outputs=AP1)
    return model
```

```python
model = Average_Operation()
model.summary()
```

```python
tf.keras.utils.plot_model(
    model, to_file='model.png', show_shapes=False, show_layer_names=True, rankdir='LR', expand_nested=False, dpi=96
)
```

```python
def Average_Operation():
    image = Input(shape=(232,232,3))
    AP = AveragePooling2D(pool_size = (2, 2))(image)
    AP1 = AveragePooling2D(pool_size = (2, 2))(AP)
    AP2 = AveragePooling2D(pool_size = (2, 2))(AP1)
    model = Model(inputs=image, outputs=AP2)
    return model
```

```python
model = Average_Operation()
model.summary()
```

```python
tf.keras.utils.plot_model(
    model, to_file='model.png', show_shapes=False, show_layer_names=True, rankdir='LR', expand_nested=False, dpi=96
)
```
