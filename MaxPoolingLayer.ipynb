```python
def Maxpoolin_Operation():
    image = Input(shape=(232,232,3))
    MP = MaxPooling2D(pool_size = (2, 2))(image)
    model = Model(inputs=image, outputs=MP)
    return model
```

```python
model = Maxpoolin_Operation()
model.summary()
```

```python
tf.keras.utils.plot_model(
    model, to_file='model.png', show_shapes=False, show_layer_names=True, rankdir='LR', expand_nested=False, dpi=96
)
```

```python
def Maxpoolin_Operation():
    image = Input(shape=(232,232,3))
    MP = MaxPooling2D(pool_size = (2, 2))(image)
    MP1 = MaxPooling2D(pool_size = (2, 2))(MP)
    MP2 = MaxPooling2D(pool_size = (2, 2))(MP1)
    model = Model(inputs=image, outputs=MP2)
    return model
```

```python
model = Maxpoolin_Operation()
model.summary()
```

```python
tf.keras.utils.plot_model(
    model, to_file='model.png', show_shapes=False, show_layer_names=True, rankdir='LR', expand_nested=False, dpi=96
)
```
